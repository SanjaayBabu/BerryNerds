```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(-100,100,10)
```

```python
x
```

```python
def c2f(temp):
    faren = (temp * 9/5) + 32
    return faren
```

```python
z = c2f(x)
```

```python
plt.plot(x,z)
plt.xlabel('Centi')
plt.ylabel('Faren')
#plt.legend('x')
plt.title('Kumar graph')
plt.grid()
plt.show()
```
